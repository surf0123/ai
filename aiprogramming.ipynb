```python
import tensorflow as tf
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
train_images = x_train / 255.0
test_images = x_test / 255.0

model = models.Sequential()
model.add(layers.Flatten(input_shape=(28, 28)))  # 784м°Ёмӣҗ ліҖнҷҳ
model.add(layers.Dense(10, activation='softmax'))  # м¶ңл Ҙмёө(лӢЁмёө)

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history=model.fit(train_images, y_train, epochs=5)

test_loss, test_acc = model.evaluate(test_images,  y_test)
print('н…ҢмҠӨнҠё м •нҷ•лҸ„:', test_acc)

loss = history.history['loss']
acc = history.history['accuracy']
epochs = range(1, len(loss)+1)

plt.plot(epochs, loss, 'b', label='Training Loss')
plt.plot(epochs, acc, 'r', label='Accuracy')
plt.xlabel('epochs')
plt.ylabel('loss/acc')
plt.show()
#0~9999
pred = model.predict(test_images[0].reshape(1,28,28,1), batch_size=1)
print("м¶”м •лҗң мҲ«мһҗ=", pred.argmax())
plt.imshow(test_images[0], cmap="Greys")
```

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()

train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

train_images, test_images = train_images / 255.0, test_images / 255.0

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
model.summary()

import matplotlib.pyplot as plt

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history=model.fit(train_images, train_labels, epochs=5)

test_loss, test_acc = model.evaluate(test_images, test_labels)
print('н…ҢмҠӨнҠё м •нҷ•лҸ„:', test_acc)

loss = history.history['loss']
acc = history.history['accuracy']
epochs = range(1, len(loss)+1)

plt.plot(epochs, loss, 'b', label='Training Loss')
plt.plot(epochs, acc, 'r', label='Accuracy')
plt.xlabel('epochs')
plt.ylabel('loss/acc')
plt.show()

pred = model.predict(test_images[0].reshape(1,28,28,1), batch_size=1)
print("м¶”м •лҗң мҲ«мһҗ=", pred.argmax())
plt.imshow(test_images[0], cmap="Greys")

```

LeNet

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

# Model / data parameters
num_classes = 10
input_shape = (28, 28, 1)

# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Scale images to the [0, 1] range
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255
# Make sure images have shape (28, 28, 1)
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)
print("x_train shape:", x_train.shape)
print(x_train.shape[0], "train samples")
print(x_test.shape[0], "test samples")

# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax"),
    ]
)
model.summary()

batch_size = 128
epochs = 15
model.compile(loss="categorical_crossentropy", optimizer   ="adam", metrics=["accuracy"])
model.fit(x_train, y_train, batch_size=batch_size, epochs   =epochs, validation_split=0.1)

score = model.evaluate(x_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])

```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

# 1) MNIST лҚ°мқҙн„° лЎңл“ң
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# LeNet мһ…л Ҙ нҳ•нғңм—җ л§һкІҢ reshape (28Г—28 вҶ’ 28Г—28Г—1)
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
x_test  = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

# 2) LeNet-5 лӘЁлҚё м •мқҳ
model = models.Sequential([
    layers.Conv2D(6, (5, 5), activation='tanh', input_shape=(28, 28, 1)),
    layers.AveragePooling2D(2,2),
    layers.Conv2D(16, (5, 5), activation='tanh'),
    layers.AveragePooling2D(2,2),
    layers.Flatten(),
    layers.Dense(120, activation='tanh'),
    layers.Dense(84, activation='tanh'),
    layers.Dense(10, activation='softmax')
])

# 3) лӘЁлҚё м»ҙнҢҢмқј л°Ҹ н•ҷмҠө
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

history = model.fit(
    x_train, y_train,
    epochs=5,
    batch_size=128,
    validation_split=0.1
)

# 4) н…ҢмҠӨнҠём…Ӣ м •нҷ•лҸ„ нҸүк°Җ
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test accuracy:", test_acc)

# 5) мҳҲмёЎ н…ҢмҠӨнҠё
sample = x_test[0].reshape(1, 28, 28, 1)
pred = model.predict(sample)
print("Predicted:", np.argmax(pred))
plt.imshow(x_test[0], cmap="Greys")
```

VGG Net

```python
from google.colab import files
from keras.applications import vgg16
from keras.preprocessing.image import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt

# 1) мқҙлҜём§Җ м—…лЎңл“ң
uploaded = files.upload()       # м—…лЎңл“ң м°Ҫмқҙ лңё
image_path = list(uploaded.keys())[0]

# 2) мқҙлҜём§Җ м „мІҳлҰ¬ н•ЁмҲҳ
def preprocess_image(image_path, target_size=None):
    img = load_img(image_path, target_size=target_size)
    img = img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = vgg16.preprocess_input(img)
    return img

# 3) VGG16 лӘЁлҚё л¶Ҳлҹ¬мҳӨкё°
base_model = vgg16.VGG16(weights='imagenet', include_top=True)
base_model.summary()


# 4) м—…лЎңл“ңн•ң мқҙлҜём§Җ мҳҲмёЎ
img_preprocessed = preprocess_image(image_path, target_size=(224, 224))
preds = base_model.predict(img_preprocessed)
print('Predicted:', vgg16.decode_predictions(preds, top=3)[0])

# 5) мқҙлҜём§Җ н‘ңмӢң
image = plt.imread(image_path)
plt.imshow(image)
plt.axis('off')
plt.show()
```

EfficientNetV2

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import EfficientNetV2S
import matplotlib.pyplot as plt
import numpy as np

# 1лҚ°мқҙн„°м…Ӣ лЎңл“ң л°Ҹ м „мІҳлҰ¬ (CIFAR-10)
print("лҚ°мқҙн„°лҘј лЎңл“ң мӨ‘мһ…лӢҲлӢӨ...")
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# EfficientNetV2S к¶ҢмһҘ мһ…л Ҙ нҒ¬кё°м—җ л§һкІҢ мЎ°м • (л©”лӘЁлҰ¬ мӮ¬м–‘м—җ л”°лқј 128~224 к¶ҢмһҘ)
IMG_SIZE = 128
CHANNELS = 3
NUM_CLASSES = 10

def preprocess_data(images, labels):
    # мқҙлҜём§Җ лҰ¬мӮ¬мқҙм§• л°Ҹ л°°м№ҳ мІҳлҰ¬мҡ© н•ЁмҲҳ
    dataset = tf.data.Dataset.from_tensor_slices((images, labels))
    dataset = dataset.map(lambda x, y: (tf.image.resize(x, (IMG_SIZE, IMG_SIZE)), y))
    dataset = dataset.shuffle(1000).batch(32).prefetch(tf.data.AUTOTUNE)
    return dataset

train_ds = preprocess_data(x_train, y_train)
test_ds = preprocess_data(x_test, y_test)

# 2. EfficientNetV2S лӘЁлҚё л№Ңл“ң
def build_efficientnet_v2(num_classes):
    # Imagenet к°ҖмӨ‘м№ҳлҘј мӮ¬мҡ©н•ҳкі  мөңмғҒлӢЁ л Ҳмқҙм–ҙ(л¶„лҘҳкё°)лҠ” м ңмҷё
    base_model = EfficientNetV2S(
        weights='imagenet',
        include_top=False,
        input_shape=(IMG_SIZE, IMG_SIZE, CHANNELS)
    )

    # нҠ№м§• м¶”м¶ңкё°(Base Model)мқҳ к°ҖмӨ‘м№ҳлҠ” кі м • (н•ҷмҠө мҶҚлҸ„ н–ҘмғҒ)
    base_model.trainable = False

    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.BatchNormalization(),
        layers.Dropout(0.3),  # кіјм Ғн•© л°©м§Җ
        layers.Dense(num_classes, activation='softmax')
    ])

    return model

model = build_efficientnet_v2(NUM_CLASSES)

model.compile( optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

print("\nн•ҷмҠөмқ„ мӢңмһ‘н•©лӢҲлӢӨ. (мҪ”лһ© GPU лӘЁл“ңлҘј к¶ҢмһҘн•©лӢҲлӢӨ)")
epochs = 5 # мӢӨмҠөмҡ©мңјлЎң 5нҡҢ м„Өм •
history = model.fit(train_ds, epochs=epochs,validation_data=test_ds)

# 5. кІ°кіј мӢңк°Ғнҷ”
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(epochs)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.title('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.title('Loss')
plt.legend()
plt.show()
```

MobileNetV3

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV3Small  # EfficientNetV2S вҶ’ MobileNetV3Small
import matplotlib.pyplot as plt
import numpy as np

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# MobileNetV3 к¶ҢмһҘ мһ…л Ҙ нҒ¬кё°м—җ л§һкІҢ мЎ°м • (128x128 к¶ҢмһҘ)
IMG_SIZE = 128
CHANNELS = 3
NUM_CLASSES = 10

def preprocess_data(images, labels):
    # MobileNetV3 м „мІҳлҰ¬ м¶”к°Җ (EfficientNetкіј лӢӨлҰ„)
    dataset = tf.data.Dataset.from_tensor_slices((images, labels))
    dataset = dataset.map(
        lambda x, y: (
            tf.keras.applications.mobilenet_v3.preprocess_input(
                tf.image.resize(tf.cast(x, tf.float32)/255.0, (IMG_SIZE, IMG_SIZE)) * 255.0
            ),
            y
        ),
        num_parallel_calls=tf.data.AUTOTUNE
    )
    dataset = dataset.shuffle(1000).batch(32).prefetch(tf.data.AUTOTUNE)
    return dataset

train_ds = preprocess_data(x_train, y_train)
test_ds = preprocess_data(x_test, y_test)

# MobileNetV3Small лӘЁлҚё л№Ңл“ң
def build_mobilenetv3(num_classes):
    # Imagenet к°ҖмӨ‘м№ҳлҘј мӮ¬мҡ©н•ҳкі  мөңмғҒлӢЁ л Ҳмқҙм–ҙ(л¶„лҘҳкё°)лҠ” м ңмҷё
    base_model = MobileNetV3Small(  # ліҖкІҪлҗң л¶Җл¶„
        weights='imagenet',
        include_top=False,
        input_shape=(IMG_SIZE, IMG_SIZE, CHANNELS)
    )

    # нҠ№м§• м¶”м¶ңкё°(Base Model)мқҳ к°ҖмӨ‘м№ҳлҠ” кі м • (н•ҷмҠө мҶҚлҸ„ н–ҘмғҒ)
    base_model.trainable = False

    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.BatchNormalization(),
        layers.Dropout(0.3),  # кіјм Ғн•© л°©м§Җ
        layers.Dense(num_classes, activation='softmax')
    ])

    return model

model = build_mobilenetv3(NUM_CLASSES)

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.summary()

print("\nн•ҷмҠөмқ„ мӢңмһ‘н•©лӢҲлӢӨ. (мҪ”лһ© GPU лӘЁл“ңлҘј к¶ҢмһҘн•©лӢҲлӢӨ)")
epochs = 5 #  5нҡҢ м„Өм •
history = model.fit(train_ds, epochs=epochs, validation_data=test_ds)

# кІ°кіј мӢңк°Ғнҷ”
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(epochs)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.title('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.title('Loss')
plt.legend()
plt.show()
```
